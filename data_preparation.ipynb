Application of Locally Weighted Scatterplot Smoothing(LOWESS)

```python
X = velocity_df['z']

data = {'z':X}

for col in velocity_df.columns[1:]:
    y = velocity_df[col]
    yest = cleveland_lowess(X, y, f = 0.5, iter = 3)
    data[col] = yest

smoothed_velocities = pd.DataFrame(data)
smoothed_velocities.head()
```

Dataset preparation for Axial Velocity prediction

```python
axial_velocity_df = make_axial_velocity_data(smoothed_velocities)
axial_velocity_df.head()
```
